```python
# import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
# import data
df = pd.read_csv('world_population.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
# check missing values
df.isnull().sum()
```

```python
# check unique values
df.nunique()
```

```python
# check duplicates
df.duplicated().sum()
```

```python
# show 10 most densely populated countries
df.sort_values(by='Density (per km²)', ascending=False).head(10)
```

```python
# correlation matrix
df.corr(numeric_only=True)
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.xticks(rotation=45)
plt.rcParams['figure.figsize'] = [16, 7]
plt.title('Correlation Matrix')
plt.show()
```

```python
# average population by continent
df.groupby('Continent').mean(numeric_only=True).sort_values(by='Density (per km²)', ascending=False)
```

```python
# population growth by continent 
df2 = df.groupby('Continent')[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].mean(numeric_only=True).sort_values(by='2022 Population', ascending=False)
df2 = df2.transpose()
plt.plot(df2)
plt.ylabel('Population')
plt.legend(df2.columns)
plt.title('Population Growth by Continent')
plt.show()
```

```python
# detecting outliers
df.boxplot(grid=False, figsize=(20, 8))
```

```python
# show african countries with highest density
africa = df[df['Continent'] == 'Africa'].sort_values(by='Density (per km²)', ascending=False)
africa
```

```python
# african countries population growth distribution
africa.boxplot(grid=False, figsize=(20, 8))
```
